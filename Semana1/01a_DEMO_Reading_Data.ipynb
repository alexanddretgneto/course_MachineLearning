## Common parameters

Existem diversos parâmetros comuns que podem ser utilizados para ler dados SQL com formatação:

- coerce_float: Tentativa de forçar números em valores float.
- parse_dates: Lista de colunas para serem analisadas como datas.
- chunksize: Número de linhas para incluir em cada fragmento.
 
 

Vamos dar uma olhada em como usar alguns desses parâmetros.


```python
query='''
SELECT Artist, Release_Year, COUNT(*) AS num_songs, AVG(PlayCount) AS avg_plays  
    FROM rock_songs
    GROUP BY Artist, Release_Year
    ORDER BY num_songs desc;
'''

# Execute the query
observations_generator = pds.read_sql(query, con, coerce_float=True, parse_dates=['Release_Year'], chunksize=5)

for index, observations in enumerate(observations_generator):
    if index < 2:
        print(f'Observations index: {index}'.format(index))
        display(observations)
```
