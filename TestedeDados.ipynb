```python
import random


def gerar_combinacoes_lotofacil(df, n_jogos=10, n_acertos=14, n_repeticoes=1):
  """
  Gera combinações para a Lotofácil com chances de acertar pelo menos 12 números,
  com a opção de gerar combinações com números repetidos.

  Parâmetros:
    df: DataFrame com os dados dos concursos da Lotofácil.
    n_jogos: Número de combinações a serem geradas (padrão: 10).
    n_acertos: Número mínimo de acertos desejado (padrão: 12).
    n_repeticoes: Número máximo de repetições de um mesmo número (padrão: 1).

  Retorno:
    Lista com n_jogos combinações de 15 números.
  """

  # Seleciona as colunas com os números sorteados
  df_numeros = df[['Bola1', 'Bola2', 'Bola3', 'Bola4','Bola5','Bola6', 'Bola7', 'Bola8', 'Bola9','Bola10','Bola11', 'Bola12', 'Bola13', 'Bola14','Bola15']]
  # Calcula a frequência de cada número
  df_frequencia = df_numeros.value_counts().sort_values(ascending=False)

  # Seleciona os números mais frequentes
  numeros_frequentes = df_frequencia.head(n_acertos).index.tolist()

  # Combinações de 15 números com base nos números frequentes
  combinacoes = []

  # Itera o número de jogos
  for i in range(n_jogos):

    # Sorteia 15 - n_acertos números aleatórios do restante dos números
    numeros_aleatorios = random.sample(list(set(range(1, 26)) - set(numeros_frequentes)), 15 - n_acertos)

    # Combina os números frequentes com os números aleatórios
    combinacao = numeros_frequentes + numeros_aleatorios

    # Permite repetições de números (se aplicável)
    if n_repeticoes > 1:
      # Converte cada elemento em 'combinacao' para uma lista antes da repetição
      combinacao = [list(x) for x in combinacao] * n_repeticoes
      # Achata a lista aninhada em uma única lista
      combinacao = sum(combinacao, [])

    # Ordena a combinação se todos os elementos forem inteiros
    if all(isinstance(x, int) for x in combinacao):
      combinacao.sort()

    # Adiciona a combinação à lista
    combinacoes.append(combinacao)

  return combinacoes

# Gera as combinações
combinacoes = gerar_combinacoes_lotofacil(df, n_jogos=1, n_acertos=14, n_repeticoes=1)


# Imprime cada combinação
for i, combinacao in enumerate(combinacoes):
  print(f"Jogo {i+1}:", combinacao)
```

```python
import pandas as pd

# Lê os dados dos concursos da Lotofácil
df = pd.read_excel("Lotofácil.xlsx")

# Define o número de jogos a serem gerados
n_jogos = 1

# Gera as combinações
combinacoes = gerar_combinacoes_lotofacil(df, n_jogos=n_jogos, n_acertos=14, n_repeticoes=1)

# Imprime cada combinação
for i, combinacao in enumerate(combinacoes):
    print(f"Jogo {i+1}:", combinacao)

# Função para calcular a probabilidade de acertar `n_acertos` números em `n_total`
def calcular_probabilidade(n_total, n_acertos):
    from math import factorial
    return factorial(n_total) / (factorial(n_acertos) * factorial(n_total - n_acertos))

# Calcula a probabilidade de acertar 14 números
probabilidade = calcular_probabilidade(25, 14)

# Imprime a probabilidade
print(f"Probabilidade de acertar 14 números: {probabilidade}")
```
