### **Date Time Transformations**


#### **Transformando a coluna de tempo 'Duração'**

Aqui, daremos uma olhada mais de perto na variável 'Duração'. Duração é o tempo que um avião leva para chegar ao seu destino. É a diferença entre 'Dep_Time' e 'Arrival_Time'. Em nosso conjunto de dados, a ‘Duração’ é expressa como uma string, em horas e minutos. Para sermos reconhecidos pelos algoritmos de aprendizado de máquina, também precisamos transformá-lo em tipo numérico.

O código abaixo irá percorrer cada registro na coluna 'Duração' e dividi-lo em horas e minutos, como duas colunas separadas adicionais. Além disso, queremos adicionar 'Duration_hours' (em minutos) à coluna 'Duration_ Minutes' para obter um tempo 'Duration_Total_mins', em minutos. A coluna de tempo de duração total será um recurso útil para qualquer tipo de análise de regressão.

```python
duration = list(data1['Duration'])
for i in range(len(duration)) :
    if len(duration[i].split()) != 2:
        if 'h' in duration[i] :
            duration[i] = duration[i].strip() + ' 0m'
        elif 'm' in duration[i] :
            duration[i] = '0h {}'.format(duration[i].strip())
dur_hours = []
dur_minutes = []  
 
for i in range(len(duration)) :
    dur_hours.append(int(duration[i].split()[0][:-1]))
    dur_minutes.append(int(duration[i].split()[1][:-1]))
     
 
data1['Duration_hours'] = dur_hours
data1['Duration_minutes'] =dur_minutes
data1.loc[:,'Duration_hours'] *= 60
data1['Duration_Total_mins']= data1['Duration_hours']+data1['Duration_minutes']
```

Imprima o quadro de dados 'data1' para ver as colunas recém-criadas.

Como você notou, três novas colunas foram criadas: 'Duration_hours', 'Duration_ Minutes' e 'Duration_Total_mins' - todos valores numéricos.

### **PCA com Scikit-Learn**

A redução da dimensionalidade faz parte do processo de extração de recursos que combina os recursos existentes para produzir recursos mais úteis. O objetivo da redução da dimensionalidade é simplificar os dados sem perder muita informação. A Análise de Componentes Principais (PCA) é um dos algoritmos de redução de dimensionalidade mais populares. Primeiro, ele identifica o hiperplano mais próximo dos dados e, em seguida, projeta os dados nele. Desta forma, alguns recursos multidimensionais são fundidos em um só.

Na parte seguinte do laboratório, usaremos a biblioteca `scikit-learn` para realizar alguns PCA em nossos dados.
Para saber mais sobre o PCA `scikit-learn`, visite esta [documentação](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term= 10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMML0232ENSkillsNetwork30654641-2022-01-01#sklearn.decomposition.PCA).

### **Escolhendo o número certo de dimensões**

Em vez de escolher arbitrariamente o número de dimensões a serem reduzidas, é mais simples escolher o número de dimensões que somam uma proporção suficientemente grande da variação, digamos 95%.

O código a seguir executa PCA sem reduzir a dimensionalidade e, em seguida, calcula o número mínimo de dimensões necessárias para preservar 95% da variação.

```python
pca = PCA()
pca.fit(x)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >=0.95) + 1
```

```python
d
```

Existem 16 componentes necessários para atender a uma variação de 95%. Portanto, poderíamos definir n_components = 16 e executar o PCA novamente. No entanto, existe uma maneira melhor, em vez de especificar o número de componentes principais que você deseja preservar, você pode definir n_components como um valor flutuante entre 0,0 e 1,0, indicando a proporção de variação que você deseja preservar.

```python
pca = PCA(n_components=0.95)
x_reduced = pca.fit_transform(x)
```

```python
x_reduced
```

Também existe uma maneira gráfica de determinar o número de componentes principais em sua análise. É traçar a variância explicada em função do número de dimensões. Geralmente haverá um cotovelo na curva, onde a variância explicada para de crescer rapidamente. Esse ponto geralmente é o ponto ideal para o número de componentes principais.

```python
import matplotlib.pyplot as plt
import numpy as np  # Assuming cumsum is a NumPy array

# Create x-axis values (number of components)
x = np.arange(1, cumsum.shape[0] + 1)  # Corrected for clarity and efficiency

# Create the area plot
plt.figure(figsize=(8, 6))  # Set appropriate figure size
plt.fill_between(x, 0, cumsum, alpha=0.7, color='skyblue')  # Area fill with transparency

# Customize labels and title
plt.xlabel("Number of Components")
plt.ylabel("Explained Variance")
plt.title("Explained Variance Ratio by Number of Components")

# Grid lines for readability
plt.grid(True)

# Display the plot
plt.show()
```
